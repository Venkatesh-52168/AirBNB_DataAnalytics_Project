```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="neighbourhood_group", order=df["neighbourhood_group"].value_counts().index)
plt.title("Number of Listings per Neighbourhood Group")
plt.xlabel("Neighbourhood Group")
plt.xticks(rotation=45)
plt.show()
```

```python
#Average price for each room type.
a = df.groupby("room_type")["price"].mean().reset_index()

sns.barplot(x="room_type", y="price", data=a)
```

```python
#Top 10 Hosts by Listing Count
top_hosts = df.groupby(["host_id", "host_name"]).size().reset_index(name="listing_count")
top_hosts_sorted = top_hosts.sort_values(by="listing_count", ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(
    data=top_hosts_sorted,
    x="listing_count",
    y="host_name",
    palette="viridis"
)

plt.title("Top 10 Hosts with the Most Listings")
plt.xlabel("Number of Listings")
plt.tight_layout()
plt.show()
```

```python
# # Bar chart: Total number of reviews by neighborhood

review_counts = df.groupby('neighbourhood')['number_of_reviews'].sum().sort_values(ascending=False).head(15)
plt.figure(figsize=(12, 6))
sns.barplot(x=review_counts.values, y=review_counts.index, palette="magma")
plt.title("Total Number of Reviews by Neighborhood (Top 15)")
plt.xlabel("Total Reviews")
plt.tight_layout()
plt.show()
```

```python
# Line plot of average availability_365 across neighborhoods
avg_availability = df.groupby('neighbourhood')['availability_365'].mean().sort_values(ascending=False).head(15)
plt.figure(figsize=(12, 6))
sns.lineplot(x=avg_availability.values, y=avg_availability.index, marker='o', color='green')
plt.title("Average Availability by Neighborhood (Top 15)")
plt.xlabel("Average Availability (Days)")
plt.tight_layout()
plt.show()
```

```python
#Box plot showing price distribution for each neighborhood group
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="neighbourhood_group", y="price", palette="pastel")
plt.title("Price Distribution by Neighborhood Group")
plt.ylim(0, 1000)
plt.xlabel("Neighborhood Group")
plt.tight_layout()
plt.show()
```

```python
# Box plot of price for each room type (to detect outliers)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="room_type", y="price", palette="Set3")
plt.title("Price Distribution by Room Type")
plt.ylim(0, 1000)
plt.xlabel("Room Type")

plt.show()
```

```python
 # Scatter plot of price vs number of reviews
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="number_of_reviews", y="price", alpha=0.5)
plt.title("Price vs. Number of Reviews")
plt.xlabel("Number of Reviews")
plt.ylim(0, 1000)
plt.tight_layout()
plt.show()
```

```python
 # Scatter plot of availability vs price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="availability_365", y="price", alpha=0.5)
plt.title("Availability vs. Price")
plt.xlabel("Availability (Days per Year)")
plt.ylim(0, 1000)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot of latitude vs longitude to show listing locations
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x="longitude", y="latitude", hue="room_type")
plt.title("Listing Locations by Latitude and Longitude")
plt.xlabel("Longitude")
plt.legend(title="Room Type")
plt.tight_layout()
plt.show()
```

```python
# Pie chart of listings by neighborhood group
plt.figure(figsize=(7, 7))
ngroup_counts = df["neighbourhood_group"].value_counts()
plt.pie(ngroup_counts, labels=ngroup_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Listing Percentage by Neighborhood Group")
plt.tight_layout()
plt.show()
```

```python
# Pie chart of listing percentages by room type
plt.figure(figsize=(7, 7))
room_counts = df["room_type"].value_counts()
plt.pie(room_counts, labels=room_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Listing Percentage by Room Type")
plt.tight_layout()
plt.show()
```

```python
 # Histogram of availability_365
plt.figure(figsize=(10, 6))
sns.histplot(df["availability_365"], bins=50)
plt.title("Histogram of Availability (Days per Year)")
plt.xlabel("Availability")
plt.tight_layout()
plt.show()
```

```python
 #Histogram of number of reviews per listing
plt.figure(figsize=(10, 6))
sns.histplot(df["number_of_reviews"], bins=50)
plt.title("Histogram of Number of Reviews")
plt.xlabel("Number of Reviews")
plt.tight_layout()
plt.show()
```

```python
 # Correlation heatmap for numerical columns
# ================================
plt.figure(figsize=(10, 6))
sns.heatmap(df[["price", "number_of_reviews", "availability_365"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
